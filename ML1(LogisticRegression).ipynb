```python
titanic_v2 = new_data_frame.drop(['PassengerId','Cabin','Name','Ticket'] , axis= 1)
titanic_v2.head()
```

```python
titanic_sex_ohe = pd.get_dummies(titanic_v2['Sex'], drop_first=False)
titanic_embarked_ohe = pd.get_dummies(titanic_v2['Embarked'], drop_first=False)
titanic_v3 = titanic_v2.drop(['Sex', 'Embarked'], axis=1).merge(titanic_sex_ohe, left_index=True, right_index=True).merge(titanic_embarked_ohe, left_index=True, right_index=True)
```

```python
titanic_v3.shape
```

```python
titanic_v3.isnull().sum()
```

```python
titanic_v3[(titanic_v3['female']==1) & (titanic_v3['Age'].isnull())]
```

```python
titanic_v3.Age.mean()
```

```python
titanic_v3['Age'] = titanic_v3['Age'].fillna(titanic_v3.Age.mean())
titanic_v3.isnull().sum()
```

```python
titanic_v3
```

```python
titanic_v3.loc[711]
```

```python
from sklearn.linear_model import  LogisticRegression
classifier = LogisticRegression(max_iter=10000)

from sklearn.model_selection import train_test_split
x = titanic_v3.drop('Survived', axis=1)
y = titanic_v3['Survived']

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=33, train_size=0.8)
classifier.fit(X_train, y_train)
```

```python
classifier.coef_
```

```python
y_pred = classifier.predict(X_test)
```

```python
from  sklearn.metrics import confusion_matrix
confusion_matrix(y_pred=y_pred, y_true=y_test)
```
